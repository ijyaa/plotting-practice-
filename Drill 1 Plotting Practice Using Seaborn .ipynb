# Singular Variable Graph ( 4 diff ways)

```python
x = df['GRE Score']
sns.distplot(x)
plt.show()
```

```python
sns.lmplot(x = 'Serial No.', y = 'GRE Score', data = df, 
           fit_reg = False, 
           hue = 'University Rating')
plt.show()
```

```python
sns.distplot(x, hist=False, rug=True)
plt.show()
```

```python
sns.kdeplot(x, shade = True)
plt.show()
```

# Two Continious Variable (3 diff ways)

```python
sns.lmplot(x = 'GRE Score', y = 'Chance of Admit ', data = df)
plt.show()
```

```python
sns.lmplot(x = 'GRE Score', y = 'Chance of Admit ', data = df,
          fit_reg = False, 
          hue = 'University Rating')
plt.show()

```

```python
sns.regplot(x='GRE Score', y= 'Chance of Admit ', data= df, x_estimator = np.mean)
plt.show()
```

## One continous and one categorical and plot six differnet ways.

```python
sns.barplot(x = 'University Rating', y = 'Chance of Admit ', data = df)
plt.show()
```

```python
sns.boxplot(x = 'University Rating', y = 'Chance of Admit ', data = df)
plt.show()
```

```python
sns.lmplot(x='University Rating', y= 'Chance of Admit ', data = df, 
          fit_reg = False, 
          hue = 'University Rating')
plt.show()
```

```python
sns.countplot(x = 'University Rating', data = df)
plt.show()
```

```python
sns.violinplot(x = 'University Rating', y = 'Chance of Admit ', data = df, palatte = 'rainbow')
plt.show()

```

```python
sns.jointplot(x='University Rating', y='GRE Score', data=df, kind = 'kde')
plt.show()
# not sure what this means....but looks cool. 
```

```python
f, axes = plt.subplots(2, 2, figsize=(7, 7), sharex=True)
sns.distplot(df['SOP'], color = 'green', ax = axes[0,0])
sns.distplot(df['LOR '], color = 'blue', ax = axes[0,1])
sns.distplot(df['CGPA'], color = 'red', ax = axes[1,0])
sns.distplot(df['University Rating'], color = 'red', ax = axes[1,1])
plt.show()
```

```python
corr = df[['University Rating','GRE Score', 'TOEFL Score', 'Chance of Admit ', 'CGPA']].corr()
```

```python
sns.heatmap(corr, annot = True)
plt.show()
```
